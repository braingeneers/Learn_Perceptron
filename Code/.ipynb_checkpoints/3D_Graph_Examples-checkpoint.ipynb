# Scatter Plot

```python
from matplotlib import pyplot as plt

import numpy as np
```

```python
np.random.seed(31)

mu = 3
n=50

x = np.random.normal(mu, 1, size=n)
y = np.random.normal(mu, 1, size=n)
z = np.random.normal(mu, 1, size=n)
```

```python
plt.rcParams['figure.figsize'] = (8,6)
```

```python
ax = plt.axes(projection='3d');
```

```python
type(ax)
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z);
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z, s=100)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

Tilting the figure

```python
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z, s=100)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.view_init(45, 215);
```

# Lineplot

```python
omega = 2

z_line = np.linspace(0, 10, 100)
x_line = np.cos(omega*z_line)
y_line = np.sin(omega*z_line)
```

```python
ax = plt.axes(projection='3d')
ax.plot3D(x_line, y_line, z_line, lw=4);
```

# 3D Surface

Example Surface: $z=50-(x^2+y^2)$

```python
def function_z(x, y):
    return 50 - (x**2 + y**2)
```

```python
N = 50  

x_values = np.linspace(-5, 5, N)
y_values = np.linspace(-5, 5, N)
```

```python
X, Y = np.meshgrid(x_values, y_values)
```

```python
X
```

```python
X.shape
```

```python
plt.scatter(X, Y);
```

```python
Z = function_z(X, Y)
```

```python
Z.shape
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z);
```

# 3D Wireframe

```python
N = 10

x_values = np.linspace(-5, 5, N)
y_values = np.linspace(-5, 5, N)
```

```python
X,Y = np.meshgrid(x_values, y_values)
Z = function_z(X, Y)
```

```python
X.shape
```

```python
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black', lw=3);
```

**Image as thumbnail**

```python
fig = plt.figure(figsize=(5, 4))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black', lw=2);
```
